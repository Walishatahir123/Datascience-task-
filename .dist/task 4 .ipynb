```python
# -------------------------------
# Week 4: Statistical & Correlation Analysis
# -------------------------------

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("house_prices1.csv")

print("=== Dataset Overview ===")
print(df.head(), "\n")

# -------------------------------
# 1️⃣ Basic Descriptive Statistics
# -------------------------------
print("=== Basic Statistics ===")
print(df.describe().T, "\n")

# Mean, Median, Mode, Variance for each column
stats = pd.DataFrame({
    "Mean": df.mean(numeric_only=True),
    "Median": df.median(numeric_only=True),
    "Mode": df.mode(numeric_only=True).iloc[0],
    "Variance": df.var(numeric_only=True)
})

print("=== Mean, Median, Mode, Variance ===")
print(stats, "\n")

# -------------------------------
# 2️⃣ Correlation Matrix
# -------------------------------
print("=== Correlation Matrix ===")
corr = df.corr(numeric_only=True)
print(corr, "\n")

# -------------------------------
# 3️⃣ Correlation with Target Variable (SalePrice)
# -------------------------------
target_corr = corr['SalePrice'].sort_values(ascending=False)
print("=== Correlation of Features with SalePrice ===")
print(target_corr, "\n")

# -------------------------------
# 4️⃣ Identify Top 3 Predictive Features
# -------------------------------
top3 = target_corr[1:4]  # exclude SalePrice itself
print("🏆 Top 3 Features Most Related to SalePrice:")
print(top3, "\n")

# -------------------------------
# 5️⃣ Optional Visualization
# -------------------------------
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# -------------------------------
# ✅ Summary
# -------------------------------
print("""
 Week 4 Completed
- Calculated mean, median, mode, variance for numeric columns
- Performed correlation analysis
- Identified top 3 features related to SalePrice
- Ready to include in report and upload to GitHub
""")
```
