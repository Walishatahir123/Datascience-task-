```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned dataset
df = pd.read_csv("house_prediction_cleaned.csv")

# Display first few rows
print(df.head())
```

```python
#  1️⃣ Bar Chart – Average Price by Location

plt.figure(figsize=(8,5))
avg_price = df.groupby('OverallQual')['SalePrice'].mean()
sns.barplot(x=avg_price.index, y=avg_price.values, palette='viridis')
plt.title('Average Sale Price by Overall Quality')
plt.xlabel('Overall Quality (1–10)')
plt.ylabel('Average Sale Price')
plt.tight_layout()
plt.show()
```

```python
# 2️⃣ Scatter Plot – Living Area vs Sale Price

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', hue='OverallQual', palette='coolwarm', s=70)
plt.title('Living Area vs Sale Price')
plt.xlabel('Above Ground Living Area (sqft)')
plt.ylabel('Sale Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['SalePrice'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Number of Houses')
plt.tight_layout()
plt.show()
```

```python
# boxplot
plt.figure(figsize=(7,5))
sns.boxplot(x='GarageCars', y='SalePrice', data=df, palette='Set2')
plt.title('Sale Price by Garage Capacity')
plt.xlabel('Number of Garage Cars')
plt.ylabel('Sale Price')
plt.tight_layout()
plt.show()
```

```python
# 5️⃣ Heatmap – Correlation Between Numerical Features
# ------------------------------------------------
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```
